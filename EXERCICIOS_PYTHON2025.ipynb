```python
#MAIOR NUMERO DE LISTA

def maior_numero(lista):
  """Encontra o maior número em uma lista."""
  if not lista:
    return None
  maior = lista[0]
  for numero in lista:
    if numero > maior:
      maior = numero
  return maior

maior_numero_lista = []  # Lista para armazenar os números

while True:
  print('** Menu de opcoes **')
  print("1. Adicionar um numero")
  print("2. Lista de numeros")
  print("3. Remover um numero")
  print("4. Maior numero da lista")  # Nova opção para encontrar o maior número
  print("5. Sair")

  opcao = input("Digite a opção desejada: ")

  if opcao == "1":
    numero = float(input("Digite o número a ser adicionado: "))
    maior_numero_lista.append(numero)
    print("Número adicionado com sucesso!")
  elif opcao == "2":
    print("Lista de números:", maior_numero_lista)
  elif opcao == "3":
    if maior_numero_lista:
      numero_remover = float(input("Digite o número a ser removido: "))
      if numero_remover in maior_numero_lista:
        maior_numero_lista.remove(numero_remover)
        print("Número removido com sucesso!")
      else:
        print("Número não encontrado na lista.")
    else:
      print("A lista está vazia.")
  elif opcao == "4":  # Opção para encontrar o maior número
    maior = maior_numero(maior_numero_lista)
    if maior is not None:
      print(f"O maior número da lista é: {maior}")
    else:
      print("A lista está vazia.")
  elif opcao == "5":
    print("Saindo do programa...")
    break
  else:
    print("Opção inválida. Tente novamente.")
```

```python
#MODOS DE LEITURA DE ARQUIVOS

'r'- leitura
'w'- escrita
'a'- append
```

```python
#ABERTURA DO ARQUIVO EM MODO DE LEITURA 'r'
arquivo = open('meu_arquivo.txt', 'r')

#LENDO O CONTEÚDO DO ARQUIVO
conteudo = arquivo.read()

print(conteudo)

#FECHAMENTO O ARQUIVO
arquivo.close()
```

```python
#ABERTURA DO ARQUIVO EM MODO DE LEITURA 'r'
arquivo = open('meu_arquivo.txt', 'r')

#LENDO LINHA A LINHA NO ARQUIVO - COM FOR
for linha in arquivo:
  print('-', linha.strip())

#FECHAMENTO O ARQUIVO
arquivo.close()
```

```python
#ABERTURA DO ARQUIVO EM MODO DE LEITURA 'r'
arquivo = open('meu_arquivo.txt', 'r')

#LENDO LINHA A LINHA - COM WHILE
linha = arquivo.readline()
while linha:
  print('-', linha.strip())
  linha = arquivo.readline()

#FECHAMENTO O ARQUIVO
arquivo.close()
```

```python
#Importando Pacotes

Para usar um pacote, você precisa importá-lo. Existem duas maneiras comuns de fazer isso:


Importando o pacote inteiro:

import nome_do_pacote



Importando partes específicas do pacote:

from nome_do_pacote import função_ou_classe


Por exemplo, se você quiser usar a função eleva_ao_quadrado de um pacote chamado meu_pacote, poderia fazer:

from meu_pacote import eleva_ao_quadrado


O Repositório Oficial de Pacotes

O Python tem um repositório oficial chamado Python Package Index (PyPI), onde você pode encontrar milhares de pacotes gratuitos. Para instalar um pacote, você usa a ferramenta PIP (Python Package Installer). Por exemplo:

pip install pandas


Versões de Pacotes

Os pacotes evoluem ao longo do tempo. Novas funcionalidades são adicionadas, bugs são corrigidos e melhorias de desempenho são feitas. Cada versão de um pacote é identificada por um número de versão, geralmente no formato MAJOR.MINOR.PATCH.


MAJOR: Indica mudanças significativas que podem quebrar a compatibilidade com versões anteriores.
MINOR: Indica a adição de novas funcionalidades sem quebrar a compatibilidade.
PATCH: Indica correções de bugs ou melhorias menores.

Por exemplo, na versão 2.2.2 do Pandas:


2 é a versão major.
2 é a versão minor.
2 é o patch.

Se você estiver usando uma versão mais recente de um pacote e encontrar problemas, pode instalar uma versão específica usando:

pip install pandas==2.2.2


NumPy: O Pacote de Cálculos Matemáticos

Agora que entendemos o que são pacotes, vamos conhecer o NumPy, um dos pacotes mais importantes para análise de dados.

O que é o NumPy?

O NumPy (Numerical Python) é um pacote que adiciona suporte para grandes arrays e matrizes multidimensionais, além de uma grande coleção de funções matemáticas de alto nível para operar esses arrays. Ele é especialmente útil para cálculos científicos e é a base para muitos outros pacotes de análise de dados.

Por Que Usar o NumPy?


Velocidade: As operações vectorizadas do NumPy são muito mais rápidas do que loops tradicionais em Python.
Eficiência: O NumPy usa menos memória para armazenar grandes conjuntos de dados.
Integração: O NumPy trabalha bem com outros pacotes de análise de dados, como o Pandas e o Matplotlib.

Instalando o NumPy

Para instalar o NumPy, use o PIP:

pip install numpy


Primeiros Passos com o NumPy

Vamos criar um exemplo simples para ver o NumPy em ação. Primeiro, importe o NumPy:

import numpy as np


Agora, vamos criar dois arrays:

array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])


E some os arrays:

soma = array1 + array2
print(soma) # Saída: [5 7 9]


Recursos Principais do NumPy


Arrays Multidimensionais: O NumPy permite criar arrays com mais de uma dimensão.
Operações Vectorizadas: Faça operações em todo o array de uma vez, sem precisar de loops.
Funções Matemáticas: O NumPy inclui funções para cálculos matemáticos avançados, como soma, multiplicação, média, desvio padrão, etc.

Pandas: O Pacote de Tratamento de Dados

Agora que conhecemos o NumPy, vamos falar sobre o Pandas, o pacote mais importante para análise de dados em Python.

O que é o Pandas?

O Pandas é um pacote que fornece estruturas de dados flexíveis e eficientes para trabalhar com dados tabulares, incluindo operações de limpeza, transformação e análise.

Por Que Usar o Pandas?


Trabalho com Dados Tabulares: O Pandas é perfeito para trabalhar com dados em formato de planilhas ou bancos de dados.
Manipulação de Dados: Filtre, ordene, agrupe e faça operações complexas em seus dados com facilidade.
Integração: O Pandas trabalha bem com outros pacotes, como o NumPy e o Matplotlib.

Instalando o Pandas

Para instalar o Pandas, use o PIP:

pip install pandas


Primeiros Passos com o Pandas

Vamos criar um exemplo simples para ver o Pandas em ação. Primeiro, importe o Pandas:

import pandas as pd


Agora, vamos criar um DataFrame (a estrutura de dados principal do Pandas):

dados = {
'Nome': ['João', 'Maria', 'Pedro'],
'Idade': [25, 31, 42]
}
df = pd.DataFrame(dados)
print(df)


Saída:

Nome Idade
0 João 25
1 Maria 31
2 Pedro 42


Recursos Principais do Pandas


DataFrames: A estrutura de dados mais usada no Pandas, semelhante a uma planilha do Excel.
Séries: Semelhantes a arrays do NumPy, mas com rótulos para os índices.
Filtros e Ordenações: Filtrar e ordenar dados é muito fácil com o Pandas.
Agrupamentos: Agrupe seus dados por categorias e faça análises agregadas.

Matplotlib: O Pacote de Visualização de Dados

Agora que conhecemos o NumPy e o Pandas, vamos falar sobre o Matplotlib, o pacote que nos permite transformar dados em gráficos.

O que é o Matplotlib?

O Matplotlib é um pacote que permite criar gráficos estáticos, animados e interativos em Python. Ele é especialmente útil para visualizar dados e apresentar resultados de forma clara.

Por Que Usar o Matplotlib?


Gráficos Profissionais: Crie gráficos de alta qualidade que podem ser usados em relatórios, apresentações e publicações.
Personalização: Personalize cores, tamanhos, rótulos e muito mais para tornar seus gráficos exatamente como você deseja.
Integração: O Matplotlib trabalha bem com o NumPy e o Pandas, permitindo que você crie gráficos diretamente a partir de seus dados.

Instalando o Matplotlib

Para instalar o Matplotlib, use o PIP:

pip install matplotlib


Primeiros Passos com o Matplotlib

Vamos criar um exemplo simples para ver o Matplotlib em ação. Primeiro, importe o Matplotlib:

import matplotlib.pyplot as plt


Agora, vamos criar um gráfico de linhas:

anos = [2018, 2019, 2020, 2021]
vendas = [100, 150, 120, 200]

plt.plot(anos, vendas)
plt.title('Vendas ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Vendas')
plt.show()


Isso criará um gráfico de linhas mostrando as vendas ao longo dos anos.

Recursos Principais do Matplotlib


Gráficos de Linhas: Mostre tendências ao longo do tempo.
Gráficos de Barras: Compare valores entre categorias.
Gráficos de Pizza: Mostre proporções de cada categoria em relação ao todo.
Personalização: Adicione títulos, rótulos, legendas e muito mais para tornar seus gráficos mais claros e informativos.
```

EXEMPLO DE VIZUALIZAÇÃO USANDO UM .CSV

```python
df = pd.read_csv("Salaries.csv")
```

```python
df.head(2)
```

```python
df = df[df["BasePay"] != "Not Provided"] #Filtragem de casos escritos "Not Provided", excluindo estes valores
```

```python
df["BasePay"] = df["BasePay"].astype(float) #transformado em float para não vir type object
```

```python
df_plots = df[["BasePay", "Year", "JobTitle"]] #considerando apenas as 3 colunas, pondo em um DataFrame: df_plots
```

```python
df_plots
```

```python
df_plots.groupby("Year").size() #função size calcula o COUNT, sem necessidade de dizer em qual coluna quero o calculo especifico
```

```python
df_plots["Year"]= df["Year"].astype(str)#modifica o tipo de dado de Year para que seja textual/string assim os numeros não aparecem, 2014.1 por exemplo, e sim o que estiver escrito no DataFrame!
qtd_por_ano = df_plots.groupby("Year").size() #salvando em um DataFrame: qtd_por_ano
```

```python
plt.figure(figsize=(8, 5)) #cria figura e delimita tamanho
plt.bar(qtd_por_ano.index, qtd_por_ano.values, color= "Skyblue")#chama função para Barras, eixo X é Index ou linhas; eixo Y são os valores Values ou seja colunas!
```

```python
plt.figure(figsize=(8, 5)) #cria figura e delimita tamanho
plt.bar(qtd_por_ano.index, qtd_por_ano.values, color= "Skyblue")#chama função para Barras, eixo X é Index ou linhas; eixo Y são os valores Values ou seja colunas!
plt.title("Quantidade por Ano")
plt.xlabel("Ano")
plt.ylabel("Quantidade")
```

```python
df_plots["Year"]= df["Year"].astype(str)
media_salario = df.groupby("Year")["BasePay"].mean()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(media_salario.index, media_salario.values)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(media_salario.index, media_salario.values)
plt.title("Salario por Ano")
plt.xlabel("Ano")
plt.ylabel("Valor")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

#Grafico Barras
ax[0].bar(qtd_por_ano.index, qtd_por_ano.values, color= "Skyblue")
ax[0].set_xlabel("Ano")
ax[0].set_ylabel("Quantidade")
ax[0].set_title("Quantidade por Ano")

#Grafico Linhas
ax[1].plot(media_salario.index, media_salario.values, color="Green")
ax[1].set_xlabel("Ano")
ax[1].set_ylabel("Valor")
ax[1].set_title("Salario por Ano")

plt.tight_layout()
```
